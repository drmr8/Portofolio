```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Analisis Pasar

```python
saham_us = pd.read_csv ("saham_us.csv")
saham_us
```

```python
highest_volume_stock = saham_us.loc[saham_us['Volume'].idxmax(), 'Kode Saham']
print('Saham dengan volume perdagangan tertinggi:', highest_volume_stock)
```

```python
saham_us['Fluktuasi Harga'] = saham_us['Harga Tertinggi'] - saham_us['Harga Terendah']
saham_us
```

```python
highest_fluctuation_stock = saham_us.loc[saham_us['Fluktuasi Harga'].idxmax(), 'Kode Saham']
print('Saham dengan fluktuasi harga tertinggi:', highest_fluctuation_stock)
```

```python
saham_us['Perubahan Harga (%)'] = ((saham_us['Harga Penutupan'] - saham_us['Harga Pembukaan']) / saham_us['Harga Pembukaan']) * 100
saham_us
```

```python
highest_closing_change = saham_us['Perubahan Harga (%)'].idxmax()

kode_saham_highest_closing_change = saham_us.loc[highest_closing_change, 'Kode Saham']
nama_perusahaan_highest_closing_change = saham_us.loc[highest_closing_change, 'Nama Perusahaan']
perubahan_harga_penutupan = saham_us.loc[highest_closing_change, 'Perubahan Harga (%)']

print('Kode Saham dengan perubahan harga penutupan tertinggi:', kode_saham_highest_closing_change)
print('Nama Perusahaan dengan perubahan harga penutupan tertinggi:', nama_perusahaan_highest_closing_change)
print('Perubahan Harga Penutupan (%):', perubahan_harga_penutupan)
```

```python
highest_volume = saham_us['Volume'].idxmax()

kode_saham_highest_volume = saham_us.loc[highest_volume, 'Kode Saham']
nama_perusahaan_highest_volume = saham_us.loc[highest_volume, 'Nama Perusahaan']
volume_perdagangan = saham_us.loc[highest_volume, 'Volume']

print('Kode Saham dengan volume perdagangan tertinggi:', kode_saham_highest_volume)
print('Nama Perusahaan dengan volume perdagangan tertinggi:', nama_perusahaan_highest_volume)
print('Volume Perdagangan:', volume_perdagangan)
```

```python
import matplotlib.pyplot as plt

plt.bar(saham_us['Nama Perusahaan'], saham_us['Fluktuasi Harga'])
plt.xlabel('Nama Perusahaan')
plt.ylabel('Fluktuasi Harga')
plt.title('Fluktuasi Harga Saham')
plt.xticks(rotation = 90)
plt.show()
```
