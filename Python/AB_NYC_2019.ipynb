```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hotel = pd.read_csv('AB_NYC_2019.csv')
hotel
```

```python
hotel.info()
```

```python
hotel.isnull().sum()
```

```python
hotel = hotel.dropna()
hotel.isnull().sum()
```

```python
hotel.head()
```

```python
average_price = hotel.groupby('room_type')['price'].mean()
average_price
```

```python
total_reviews = hotel.groupby('neighbourhood_group')['number_of_reviews'].sum()
total_reviews = total_reviews.sort_values(ascending=False)

plt.bar(total_reviews.index, total_reviews.values)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Reviews')
plt.title('Total Reviews per Neighbourhood Group')
plt.xticks(rotation=45)
plt.show()
```

```python
top_hosts = hotel['host_name'].value_counts().head(5)
top_hosts
```

```python
average_availability = hotel.groupby('neighbourhood')['availability_365'].mean()
average_availability
```

```python
neighborhood_availability = hotel.groupby('neighbourhood')['availability_365'].sum()
neighborhood_availability = neighborhood_availability.sort_values(ascending=False).head(5)
plt.bar(neighborhood_availability.index, neighborhood_availability.values)
plt.xlabel('Neighbourhood')
plt.ylabel('Availability')
plt.title('Availability per Neighbourhood')
plt.xticks(rotation=90)
plt.show()
```

```python
room_type_reviews = hotel.groupby('room_type')['number_of_reviews'].sum()
plt.bar(room_type_reviews.index, room_type_reviews.values)
plt.xlabel('Room Type')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews per Room Type')
plt.bar_label(plt.bar(room_type_reviews.index, room_type_reviews.values), fmt='%d')
plt.show()
```

```python
room_type_counts = hotel['room_type'].value_counts()

plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%',
        wedgeprops={'linewidth': 2, 'edgecolor': 'white'})
plt.title('Percentage of Room Types')
plt.show()
```
